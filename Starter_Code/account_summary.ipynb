# Plaid Access Token

In this section, you will use the plaid-python api to generate the correct authentication tokens to access data in the free developer Sandbox. This mimics how you might connect to your own account or a customer account, but due to privacy issues, this homework will only require connecting to and analyzing the fake data from the developer sandbox that Plaid provides. 

Complete the following steps to generate an access token:
1. Create a client to connect to plaid
2. Use the client to generate a public token and request the following items: 
['transactions', 'income', 'assets']
3. Exchange the public token for an access token
4. Test the access token by requesting and printing the available test accounts

### 1. Create a client to connect to plaid

```python
INSTITUTION_ID = "ins_109508"
```

```python
#creating client connection and passing all the plaid info
client_connection = plaid.Client(client_id=PLAID_CLIENT_ID, secret=PLAID_SECRET, public_key=PLAID_PUBLIC_KEY, environment=PLAID_ENV, api_version='2019-05-29')
```

### 2. Generate a public token

```python
client_connection.Institutions.get(2)

# Create public token to be exchanged for institution access token
create_tkn_response = client_connection.Sandbox.public_token.create(INSTITUTION_ID,['transactions','income','assets'])

```

### 3. Exchange the public token for an access token

```python
# Exchange public token for access token
exchange_response = client_connection.Item.public_token.exchange(create_tkn_response['public_token'])

# Store access token as variable
access_token = exchange_response['access_token']
```

### 4. Fetch Accounts

```python
# Get accounts associated with institution
accounts = client_connection.Accounts.get(access_token)
pretty_print_response(accounts)
```

# Account Transactions with Plaid

In this section, you will use the Plaid Python SDK to connect to the Developer Sandbox account and grab a list of transactions. You will need to complete the following steps:


1. Use the access token to fetch the transactions for the last 90 days
2. Print the categories for each transaction type
3. Create a new DataFrame using the following fields from the JSON transaction data: `date, name, amount, category`. (For categories with more than one label, just use the first category label in the list)
4. Convert the data types to the appropriate types (i.e. datetimeindex for the date and float for the amount)

### 1. Fetch the Transactions for the last 90 days

```python
# Get transactions for institution for specific date range
start_date = '{:%Y-%m-%d}'.format(datetime.datetime.now() + datetime.timedelta(-90))
end_date = '{:%Y-%m-%d}'.format(datetime.datetime.now())

# Get transactions for date range
transaction_response = client_connection.Transactions.get(access_token,start_date,end_date)

# Print JSON output
pretty_print_response(transaction_response['transactions'][:2])
```

### 2. Print the categories for each transaction

```python
all_transactions = transaction_response['transactions']

for cat in all_transactions:
    print(cat['category'])
```

### 3. Create a new DataFrame using the following fields from the JSON transaction data: date, name, amount, category. 

(For categories with more than one label, just use the first category label in the list)

```python
transactionDetails = pd.DataFrame(columns=['Date', 'Name','Amount','Category'])

details = [] #intiate an empty list to store all the rows
for row in all_transactions:
    details.append(pd.Series([row['date'], row['name'], row['amount'], row['category'][0]],  index=transactionDetails.columns))

transactionDetails = transactionDetails.append(details, ignore_index = True)      
transactionDetails
```

### 4. Convert the data types to the appropriate types 

(i.e. datetimeindex for the date and float for the amount)

```python
# Change the data type of Date to datetime, make it the index and sorty

if transactionDetails.index.name != 'Date':        
    transactionDetails['Date'] = pd.to_datetime(transactionDetails['Date'], infer_datetime_format=True)
    transactionDetails.set_index('Date', inplace=True)
transactionDetails.sort_index(inplace=True)

# Change the data type of Amount to float
transactionDetails['Amount'] = transactionDetails['Amount'].astype('float')
transactionDetails.dtypes
```

### Calculate the expenses per category

```python
#Calculate the total spending per category and print the results (Hint: groupby or count transactions per category)

# Based on above assumptions we have to eliminate the transaction amount less than 0 because its not a spending
# its a part of income

spending =transactionDetails[transactionDetails['Amount'] > 0]
total_spending = spending.groupby('Category').sum()
print(f'Total Spending per Category: \n {total_spending}')
```

```python
#Spending Categories Pie Chart

total_transaction_percategory = spending.groupby('Category').count()[['Amount']]
total_spending.plot.pie(subplots=True, figsize=(10, 30))
```

### Calculate the expenses per month

```python
#Calculate the expenses per month

exppm = spending.groupby(pd.Grouper(freq='M')).sum()
exppm.index = exppm.index.strftime('%B')
exppm
```

```python
#Plot the total expenses per month
ti="Total Spending Per Month"
ax = exppm.plot.bar(title=ti,colormap='Spectral', rot=0)
ax.set(xlabel='Months', ylabel='Total Expenses')
```
